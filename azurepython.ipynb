```python
from azure.storage.blob import BlobServiceClient

# Remplacez "AZURE_STORAGE_CONNECTION_STRING" par votre propre chaîne de connexion
connection_string = ""

# Créez un client BlobService
blob_service_client = BlobServiceClient.from_connection_string(connection_string)

# Nom du conteneur et nom du blob
container_name = "image"
blob_name = "téléchargement (4).png"

# Obtenez le blob
blob_client = blob_service_client.get_blob_client(container=container_name, blob=blob_name)

# Téléchargez le blob localement
with open(blob_name, "wb") as my_blob:
    download_stream = blob_client.download_blob()
    my_blob.write(download_stream.read())

# Affichez l'image localement
from IPython.display import Image
Image(blob_name)

```

```python
from azure.storage.blob import BlobServiceClient
import pandas as pd
import io

# Remplacez "AZURE_STORAGE_CONNECTION_STRING" par votre propre chaîne de connexion


# Créez un client BlobService
blob_service_client = BlobServiceClient.from_connection_string(connection_string)

# Nom du conteneur et nom du blob (fichier CSV)
container_name = "dataset"
blob_name = "IR007_0.csv"

# Obtenez le blob
blob_client = blob_service_client.get_blob_client(container=container_name, blob=blob_name)

# Téléchargez le contenu du blob
downloaded_blob = blob_client.download_blob()
csv_data = downloaded_blob.readall()

# Utilisez pandas pour lire le contenu du fichier CSV
df = pd.read_csv(io.BytesIO(csv_data))

# Affichez le DataFrame
print(df)
```

```python
from azure.storage.blob import BlobServiceClient
from IPython.display import Video
import base64

# Remplacez "AZURE_STORAGE_CONNECTION_STRING" par votre propre chaîne de connexion

# Créez un client BlobService
blob_service_client = BlobServiceClient.from_connection_string(connection_string)

# Nom du conteneur et nom du blob (fichier vidéo)
container_name = "video"
blob_name = "1 minute funny videos-zBjJUV-lzHo.mp4"

# Obtenez le blob
blob_client = blob_service_client.get_blob_client(container=container_name, blob=blob_name)

# Téléchargez la vidéo localement
downloaded_blob = blob_client.download_blob()
video_data = downloaded_blob.readall()

# Encodez les données de la vidéo en base64
video_base64 = base64.b64encode(video_data).decode('utf-8')

# Créez l'URL de la vidéo
video_url = "data:video/mp4;base64," + video_base64

# Affichez la vidéo
Video(video_url)
```
